```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read img 
img = cv2.imread('images.jpg')
```

```python
plt.imshow(img)
```

```python
# Convert into color
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# 1
## Transform the image in the +x direction by 25%, and create an image

```python
# build a transformation matrix
matrix1 = [[1, 0, 25], #x 
          [0, 1, 0]] #y
t1 = np.float32(matrix1)
# get the sizes
h1, w1= img.shape[:2]
# transform
img = cv2.warpAffine(img, t1, (w1, h1))
```

```python
# plot
fig, ax = plt.subplots(1, figsize=(10,6))
ax.axis('off')   
plt.imshow(img)
```

# 2
## Transform the image in the +y direction by 25%, and create an image

```python
matrix2 = [[1, 0, 0], #x 
          [0, 1, 25]] #y
t2 = np.float32(matrix2)
h2, w2 = img.shape[:2]
img = cv2.warpAffine(img, t2, (w2, h2))
```

```python
fig, ax = plt.subplots(1, figsize=(10,6))
ax.axis('off')   
plt.imshow(img)
```

## we need to define a 2 x 3 matrix called an affine transformation matrix:
    
M=[[1,0,t_[x],
  0,1,t_[y]]

### Negative values for the t_[x] value will shift the image to the left
### Positive values for  t_[x] shifts the image to the right
### Negative values for  t_[y] shifts the image up
### Positive values for  t_[y] will shift the image down
